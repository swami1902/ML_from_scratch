```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'C:\Users\Aryan Swami\Documents\Jupyter Notebook\data.csv')
```

```python
df = df.drop(['id','Unnamed: 32'],axis=1)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': '0', 'B': '1'})
```

```python
# Seperating Feature and Target
##Note:whn we drop column we put axis = 1 and whn we drop a column we put axis=0
X = df.drop(columns= 'diagnosis', axis = 1)
y = df['diagnosis']
```

```python
#Splittin the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.3, random_state=420)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
#Model Training
model = LogisticRegression()
```

```python
#training the model using training data

model.fit(X_train,y_train)
```

```python
#Model Evaluation (Using accuracy_score)
X_train_prediction = model.predict(X_train)

#here we are comparing Xtrain data and y_train data to chk accuracy of our model
training_data_accuracy = accuracy_score(y_train,X_train_prediction)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
#Model Evaluation (Using accuracy_score) ON TEST DATA
X_test_prediction = model.predict(X_test)

#here we are comparing X_test data and y_test data to chk accuracy of our model
testing_data_accuracy = accuracy_score(y_test,X_test_prediction)
```

print('Accuracy on testing data:', testing_data_accuracy)
